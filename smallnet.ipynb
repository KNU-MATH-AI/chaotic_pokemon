```python
# from google.colab import drive
# drive.mount("/content/drive")

import pandas as pd

data1 = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/data1.csv")

y_train = data1.f.to_numpy()
x_train = data1[["x","y","z"]].to_numpy()
```

# 새 섹션

```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential()
model.add(layers.Dense(32, kernel_initializer='uniform', input_shape=(3,)))
model.add(layers.Activation('softmax'))
model.add(layers.Dense(16))
model.add(layers.Activation('relu'))
model.add(layers.Dense(1, kernel_initializer='uniform'))

loss_fn = keras.losses.MeanSquaredError(reduction="auto", name="mean_squared_error")
model.compile(loss=loss_fn, optimizer='adam')

data = np.arange(60).reshape(3, 20)

model.summary()

example_batch = x_train[:10]
example_result = model.predict(example_batch)
example_result
```

```python
class PrintDot(keras.callbacks.Callback):
  def on_epoch_end(self, epoch, logs):
    if epoch % 100 == 0: print('')
    print('.', end='')

EPOCHS = 1000

history = model.fit(
  x_train, y_train,
  epochs=EPOCHS, validation_split = 0.2, verbose=0,
  callbacks=[PrintDot()])
```

```python
hist = pd.DataFrame(history.history)
hist['epoch'] = history.epoch
hist.tail()

model.predict(example_batch)
```
