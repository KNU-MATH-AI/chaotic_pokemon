```python
import tensorflow as tf
import numpy as np
import pandas as pd

from keras.models import Sequential, Model
from keras.layers import Dense, Input, Flatten, BatchNormalization
from keras.optimizers import Adam

csv_data = pd.read_csv("data1.csv")

data = csv_data.to_numpy()

N = 90 #traing number
n = 100-N

x_train = data[0:N,1:4]
x_train = np.reshape(x_train,(N,1,3))

y_train = data[0:N, 0]
y_train = np.reshape(y_train,(N,1))

x_test = data[N:100,1:4]
x_test = np.reshape(x_test,(n,1,3))

y_test = data[N:100, 0]
y_test = np.reshape(y_test,(n,1))
```

```python
print(np.shape(x_train))
print(np.shape(y_train))
print(np.shape(x_test))
print(np.shape(y_test))
```

# architecture

```python
input_layer = Input((1,3))

x = Flatten()(input_layer)

x = Dense(10, activation = 'relu')(x)

x = Dense(50, activation = 'relu')(x)

x = Dense(100, activation = 'relu')(x)

x = Dense(50, activation = 'relu')(x)

x = Dense(10, activation = 'relu')(x)

output_layer = Dense(1, activation='linear')(x)

model = Model(input_layer, output_layer)

model.summary()
```

```python
opt = Adam(lr=0.005)
model.compile(loss='mean_squared_error', optimizer=opt)
```

```python
model.fit(x_train
          , y_train
          , batch_size=5
          , epochs=100
          , shuffle=True)

model.evaluate(x_test, y_test, batch_size=1)
```
